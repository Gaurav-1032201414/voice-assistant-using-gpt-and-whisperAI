```python
import pandas as pd
import json
```

```python
# df = pd.json_normalize(json.loads("afdb.jsonl"))

# data = None
# with open("afdb.jsonl") as jsonfile:
#     data = json.load(jsonfile)

Infra_projects = [json.loads(line) for line in open("afdb.jsonl", "r")]
```

```python
df = pd.json_normalize(Infra_projects)
```

```python
df.head()
```

```python
df.info()
```

```python
prompt_cleaning = df["prompt"].str.split()
```

```python
project_code = [prompt_cleaning[i][0] for i in range(0, 1480)]
print(project_code)
```

```python
df["project code"] = project_code
```

```python
print(prompt_cleaning[0])
```

```python
project_status = [prompt_cleaning[i][len(prompt_cleaning[i])-2] for i in range(0, 1480)]
print(project_status)
```

```python
df["project_status"] = project_status
```

```python
df.head()
```

```python
project_name = []
for i in range(0, 1480):
    index_temp = prompt_cleaning[i].index("which")
    name_temp = [prompt_cleaning[i][j] for j in range(2, index_temp)]
    project_name.append(" ".join(name_temp))

print(project_name)
```

```python
df["project name"] = project_name
```

```python
df.head()
```

```python
completion_cleaning = df["completion"].str.split()
```

```python
approval_date = [completion_cleaning[i][-1] for i in range(0, 1480)]
print(approval_date)
```

```python
df["Approval date"] = approval_date
```

```python
df.head()
```

```python
df.describe()
```

```python
def project_code_questions(code):
    return "Details regarding project " + code
```

```python
code_questions = project_code_questions(df["project code"])
print(code_questions)
```

```python
df["Code Questions"] = code_questions
```

```python
df.head()
```

```python
def approval_date_questions(date):
    return "Projects approval after/before " + date

date_questions = approval_date_questions(df["Approval date"])
print(date_questions)
```

```python
project_data_csv = df.to_csv('Project_data_cleaned.csv', index = True)
```
